```python
import pandas as pd
```

```python
df = pd.read_csv("2016-building-energy-benchmarking.csv")
```

```python
df.head()
```

```python
#columns to drop: City, State, PropertyGFABuilding(s), TaxParcelIdentificationNumber
```

```python
df.drop(["City", "State", "PropertyGFABuilding(s)", "TaxParcelIdentificationNumber"], axis=1, inplace=True)
```

```python
df.head()
```

```python
#kBtu to kWh
# columns to convert: SiteEnergyUse(kBtu) - 29, SiteEnergyUseWN(kBtu) - 31, SteamUse(kBtu) - 33, NaturalGas(kBtu) - 39
```

```python
const = 3.4121416
```

```python
df.insert(loc=30, column='SiteEnergyUse(kWh)', value=df['SiteEnergyUse(kBtu)'] * const)
df.insert(loc=32, column='SiteEnergyUseWN(kWh)', value=df['SiteEnergyUseWN(kBtu)'] * const)
df.insert(loc=34, column='SteamUse(kWh)', value=df['SteamUse(kBtu)'] * const)
df.insert(loc=39, column='NaturalGas(kWh)', value=df['NaturalGas(kBtu)'] * const)
```

```python
df.info()
```

```python
#NaN rows to delete
df.dropna(subset=['SiteEnergyUse(kBtu)', 'SiteEnergyUse(kWh)', 'SiteEnergyUseWN(kBtu)',
                        'SiteEnergyUseWN(kWh)', 'SteamUse(kBtu)', 'SteamUse(kWh)',
                        'Electricity(kWh)', 'Electricity(kBtu)', 'NaturalGas(therms)'],
                inplace=True)
```

```python
df.select_dtypes(include=['float64']).info()
```

```python
#converting kWh to int -  SiteEnergyUse(kWh), SiteEnergyUseWN(kWh), SteamUse(kWh) , Electricity(kWh) , NaturalGas(kWh)
```

```python
convert_to_int = ["SiteEnergyUse(kWh)", "SiteEnergyUseWN(kWh)", "SteamUse(kWh)", "Electricity(kWh)", "NaturalGas(kWh)"]
```

```python
for column in convert_to_int:
    df[column] = df[column].astype(int)
```

```python
df.select_dtypes(include=['int']).info()
```

```python
#rows with 0 values to delete
```

```python
df = df[df['SiteEnergyUse(kBtu)'] != 0]
```

```python
df.to_csv("2016-building-energy-benchmarking-clean.csv", index=False)
```
